```python
#Abre o site e acessa endereço
sleep(10)
url = 'http://visual.ic.uff.br/dmi/prontuario/home.php'
driver = webdriver.Chrome(executable_path=r'./chromedriver.exe')
driver.get(url)
sleep(30)
driver.maximize_window()
sleep(10)

#Dados de login
username = driver.find_element_by_xpath(r'//*[@id="usuario"]')
password = driver.find_element_by_id("password")
#username.send_keys("jeanmarques")
#password.send_keys("NPXsazLNemCP@48")
#
username.send_keys(usuario)
password.send_keys(senha)
#
login_attempt = driver.find_element_by_xpath('/html/body/nav/div/div/div/ul/li[5]/form/button')
login_attempt.click()
#
sleep(2)

#Clique na lista de pacientes                              
driver.find_element_by_xpath('/html/body/div[2]/div/div[2]/div[3]/div[1]/a').click()



#Recupera tabela
tabelas = []
#Grava pag 1
tabelas.append(driver.find_element_by_xpath('//*[@id="mytable"]/tbody').text.split('\n'))
print('gravou pag 1')

#Clica no botão next
driver.find_element_by_xpath('/html/body/div[2]/div/div[2]/div[2]/ul/li[12]/a').click()
#Grava pag 2
tabelas.append(driver.find_element_by_xpath('//*[@id="mytable"]/tbody').text.split('\n'))
print('gravou pag 2')

#Clica no botão next
driver.find_element_by_xpath('/html/body/div[2]/div/div[2]/div[2]/ul/li[13]/a').click()
#Grava pag 3                  
tabelas.append(driver.find_element_by_xpath('//*[@id="mytable"]/tbody').text.split('\n'))
print('gravou pag 3')

#Clica no botão next
driver.find_element_by_xpath('/html/body/div[2]/div/div[2]/div[2]/ul/li[13]/a').click()
#Grava pag 4
tabelas.append(driver.find_element_by_xpath('//*[@id="mytable"]/tbody').text.split('\n'))
print('gravou pag 4')

#Clica no botão next
driver.find_element_by_xpath('/html/body/div[2]/div/div[2]/div[2]/ul/li[13]/a').click()
#Grava pag 5
tabelas.append(driver.find_element_by_xpath('//*[@id="mytable"]/tbody').text.split('\n'))
print('gravou pag 5')

#Clica no botão next
driver.find_element_by_xpath('/html/body/div[2]/div/div[2]/div[2]/ul/li[13]/a').click()
#Grava pag 6
tabelas.append(driver.find_element_by_xpath('//*[@id="mytable"]/tbody').text.split('\n'))
print('gravou pag 6')

#Clica no botão next
driver.find_element_by_xpath('/html/body/div[2]/div/div[2]/div[2]/ul/li[13]/a').click()
#Grava pag 7
tabelas.append(driver.find_element_by_xpath('//*[@id="mytable"]/tbody').text.split('\n'))
print('gravou pag 7')

#Clica no botão next
driver.find_element_by_xpath('/html/body/div[2]/div/div[2]/div[2]/ul/li[13]/a').click()
#Grava pag 8
tabelas.append(driver.find_element_by_xpath('//*[@id="mytable"]/tbody').text.split('\n'))
print('gravou pag 8')

#Clica no botão next
driver.find_element_by_xpath('/html/body/div[2]/div/div[2]/div[2]/ul/li[13]/a').click()
#Grava pag 9
tabelas.append(driver.find_element_by_xpath('//*[@id="mytable"]/tbody').text.split('\n'))
print('gravou pag 9')

#Clica no botão next
driver.find_element_by_xpath('/html/body/div[2]/div/div[2]/div[2]/ul/li[13]/a').click()
#Grava pag 10
tabelas.append(driver.find_element_by_xpath('//*[@id="mytable"]/tbody').text.split('\n'))
print('gravou pag 10')

#Clica no botão next
driver.find_element_by_xpath('/html/body/div[2]/div/div[2]/div[2]/ul/li[13]/a').click()
#Grava pag 11
tabelas.append(driver.find_element_by_xpath('//*[@id="mytable"]/tbody').text.split('\n'))
print('gravou pag 11')
##
#Cria colunas do dataframe
colunas = tabelas[0][0].split()
colunas.pop()
##
## Faz tratamento dos dados da tabela #######################################################################
reg = []
for tabela in tabelas:
    for i, linha in enumerate(tabela):
        if i != 0:
            reg.append(linha.lower().replace(', ', '-').split())

##
repetidos = []
ID = []
Records = [] 
Name = [] 
Age = [] 
Exams = [] 
Diagnosis = [] 
#
for i, item in enumerate(reg):
    if (int(item[0])) not in repetidos:
        repetidos.append(int(item[0]))
        ID.append(item[0])
        Records.append(item[1])
        Name.append(item[2])
        Age.append(item[3])
        Exams.append(item[4])
        Diagnosis.append(item[5])
##
####### tranforma dados das tabelas em dataframe ######################################################
df = pd.DataFrame(list(zip(ID,Records,Name,Age,Exams,Diagnosis)), columns = colunas)
#
#######################################################################################################
#Entra nos links de detalhamentos para pegar detalhamentos como dieta, exames, sintomas etc..

string_link = 'http://visual.ic.uff.br/dmi/prontuario/details.php?id='
#   
string_comment_ini = '//*[@id="exame-termico'
string_comment_fim = '"]/div/div[1]'
#
comments = []
comments_id = []
#
caract_id = []
caract_pac = []
#
for value in df.ID:
    link_paciente = (string_link + value)
    link_comment = (string_comment_ini + value + string_comment_fim)
    driver.get(link_paciente)
#    
    comments.append(driver.find_element_by_xpath('/html/body/div[2]/div/div[2]/section/div/div[1]').text.split('\n'))
    comments_id.append(value)
#
    caract_pac.append(driver.find_element_by_xpath('/html/body/div[2]/div/div[2]/div[1]/div[2]').text.split('\n'))
    caract_id.append(value)
#
## Faz tratamento dos dados das caracteristicas do paciente ###########################################
#
data_registro = []
estado_civil = []
raca = []
for lista in caract_pac:
    for i, item in enumerate(lista):
        if i == 3:                #descarta id, volunteer e idade
            #Trata a data
            start = item.find('at ') + 3
            end = item.find(' (y-m-d)', start)
            data_registro.append(item[start:end])
        elif i == 4:
            #Trata Estado civil
            start = item.find('Marital status: ') + 16
            end = item.find('. ', start)
            estado_civil.append(item[start:end])
            
            #Trata Raça
            start = item.find('Race: ') + 6
            raca.append(item[start:])
        else:
            continue
#            
####### tranforma dados das características em dataframe ##############################################
df_caract = pd.DataFrame(list(zip(caract_id,data_registro,estado_civil,raca)), columns = ['ID','Data_registro','Estado_civil', 'Raca'])
#    
#######################################################################################################
#Transforma detalhes em dataframe
df_comments = pd.DataFrame(list(zip(comments_id,comments)), columns = ['ID','Comments'])
#######################################################################################################
#Grava dataframes como csv
#
df.to_csv('tabulares', encoding='utf-8', index=False)
df_comments.to_csv('detalhes', encoding='utf-8', index=False)
df_caract.to_csv('caracteristicas', encoding='utf-8', index=False)
#######################################################################################################
#
#Começa busca pelas imagens com base in ID do dataframe

#Define url base da foto
link_base_imagem = 'http://visual.ic.uff.br/dmi/bancovl/'
#Para link total falta o id do paciente e o nome do arquivo

for i, id in enumerate(df.ID):
    link_paciente = (string_link + id)
    driver.get(link_paciente)
    lista_fotos = driver.find_element_by_xpath('/html/body/div[2]/div/div[2]/section/div/div[4]/ul').text.split('\n')
    tamanho = len(lista_fotos)
    #criar pasta para novas fotos
    #Definir o nome da pasta de acordo com id
    prefix = 'PACIENTE_'
    nome_pasta = (prefix + id.zfill(4))
    if not os.path.exists(nome_pasta):
        os.makedirs(nome_pasta)
    #Definie restante do link para a imagem
    paciente_id = (id.zfill(4))
    #
    #Itera sobre a lista de imagens
    for z, imagem in enumerate(lista_fotos):
        link_final_img = (link_base_imagem + paciente_id + '/' +lista_fotos[z][:26] + 'jpg')
        #
        #Request da imagem
        sleep(1)
        r = requests.get(link_final_img)
        sleep(2)
        nome_arq = ('PAC_' + paciente_id + '_IMG_' + str(z).zfill(4))
        caminho_arquvivo = ('./' + nome_pasta + '/' + nome_arq + '.jpg')
        with open(caminho_arquvivo, 'wb') as imagem:
            imagem.write(r.content)
#######################################################################################################
#Fecha conexão
#
driver.close()
```
